```python
import pandas as pd
import numpy as np
```

```python
df_daily = pd.read_csv('daily-chaps.csv')
```

```python
df_daily.head()
```

```python
df_daily_long = pd.melt(df_daily, id_vars=["Date"])
df_daily_long['geography_code'] = 'K02000001'
df_daily_long['geography_label'] = 'United Kingdom'
df_daily_long['period_type'] = 'day'
df_daily_long['Date'] = pd.to_datetime(df_daily_long['Date']).dt.strftime('%Y-%m-%d')
df_daily_long = df_daily_long.rename(columns={"Date": "period_code","variable":"measure", "value":"observation"})
df_daily_long['unit']= 'number'
df_daily_long['period_label']= df_daily_long['period_code']
df_daily_long['seasonal_adjustment'] = "NSA"
df_daily_long = df_daily_long[["period_code","period_label","period_type","geography_code","geography_label", "observation","measure", "unit","seasonal_adjustment"]]
```

```python
df_daily_long.head()
```

```python
conditions = [df_daily_long['measure'] == "Aggregate",
              df_daily_long['measure'] == "Delayable",
              df_daily_long['measure'] == "Social",
              df_daily_long['measure'] == "Staple",
              df_daily_long['measure'] == "Work Related", ]
choices = ['Aggregate UK spending on credit and debit cards', 'Delayable UK spending on credit and debit cards','Social UK spending on credit and debit cards', 'Staple UK spending on credit and debit cards', 'Work Related UK spending on credit and debit cards']

df_daily_long['measure'] = np.select(conditions, choices, default=0)
```

```python
df_daily_long.head()
```
